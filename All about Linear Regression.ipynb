# Multicollinearity in Linear Regression

```python
df = pd.read_csv("Advertising.csv",index_col = 0)
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1].to_frame()  # to convert series to dataframe

y
```

```python
import statsmodels.api as sm

#fit OLS model on X
X = sm.add_constant(X)  #adds const to all rows
X
```

```python
model = sm.OLS(y, X).fit() #endog is output and exog is input value
```

```python
model.summary() #const ic the c value
```

```python
#finding correlation between independent features
X.iloc[:,1:].corr()
```

correlation is less than 0.5 in this table it means that there is no multi-collinearity between your independent features hence we can use this model further

## Another example for multi-collinearity

```python
df = pd.read_csv('Salary_Data.csv')
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1].to_frame()

y
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model
```

```python
model.summary()
```

if we have a high number of std error then there's a huge multicollinearity

```python
X.iloc[:,1:].corr()
```

we have more correlation between both the features and one feature needs to be dropped because the correlation value is becoming same. so there are two remedies:

1. you dont do anything
2. compare the p-value of both the features if one feature has higher p-value then drop that feature so in this case we have age as a higher p-value so we will drop beacause the years of experience is covering 98% of the properties of age so we will train the model based on that.
