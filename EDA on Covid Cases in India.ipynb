```python
import pandas as pd

df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
#since we dont want the last row and the 1st column so

df = df.iloc[:-1,1:]
df
```

```python
import matplotlib.pyplot as plt
import numpy as np


#storing values for visualizing data
total_states = np.arange(len(df['state_name']))
total_states
```

### Total positive cases based on states in India

```python
from matplotlib.pyplot import figure
import seaborn as sns

sns.barplot(data = df, x= 'positive', y= total_states)
figure(figsize=(20,20), dpi=80)

plt.barh(total_states, df['positive'], align = 'center', alpha = 0.5)

plt.yticks(total_states , df['state_name'])
plt.xlabel('Positive no of cases')
plt.title('Corona Virus Cases')
plt.xlim(1,max(df['positive']) + 100)
```

### Total no of active cases

```python
sns.barplot(data = df, x= 'active', y= total_states)
figure(figsize=(20,20), dpi=80)

plt.barh(total_states, df['active'], align = 'center', alpha = 0.5)

plt.yticks(total_states , df['state_name'])
plt.xlabel('Active no of cases')
plt.title('Corona Virus Cases')
plt.xlim(1,max(df['active']) + 100)
```

### Death cases

```python
sns.barplot(data = df, x= 'death', y= total_states)
figure(figsize=(20,20), dpi=80)

plt.barh(total_states, df['death'], align = 'center', alpha = 0.5)

plt.yticks(total_states , df['state_name'])
plt.xlabel('Positive no of cases')
plt.title('Corona Virus Cases')
plt.xlim(1,max(df['positive']) + 10)
```
