# CSV 

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'  #basic csv creation with \n indicating next line
              'x,y,1\n'
               'a,b,2\n'
                'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data)) #to convert into dataframe from string memory buffer we will use stringIO
```

```python
pd.read_csv(StringIO(data), usecols=['col1','col2']) #to read only specific columns
```

```python
data = ('a,b,c,d\n'
       '1,2,3,4\n'
       '5,6,7,8\n'
       '9,10,11,12')
```

```python
df=pd.read_csv(StringIO(data),dtype=object) #for converting string type
df
```

```python
df['a']
```

```python
df=pd.read_csv(StringIO(data),dtype=float) #for float type conversion
df
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':float}) # to specify specific columns at different datatypes we will use dictionary like if i want to make the c column as float i will use c:float
df
```

```python
df.dtypes #to check all the data types of all columns
```

```python
df['a'][2] #(row x column) check for value 
```

```python
data = ('index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0) #to make the first column as index we can make the others columns as well
```

```python
data = ('a,b,c\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False) #to handle NAN values and preventing them from being the index we will make the index column as False and a values will be written under the a column b values under b and c values under c hence there will be no empty column NAN left
```

```python
pd.read_csv(StringIO(data),usecols=['a','b'],index_col=0)
```

#  Quoting escape sequences in NLP

```python
data = 'a,b\n"hello, \\"Bob\\",nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\') #to skip the \\ character from the dataframe
```

# Read JSON to CSV

```python
data = [
    {
        "employee name": "James",
        "email": "james@gmail.com",
        "job profile": "Team Lead"
    },
    {
        "employee name": "Emily",
        "email": "emily@example.com",
        "job profile": "Project Manager"
    },
    {
        "employee name": "John",
        "email": "john@example.com",
        "job profile": "Software Engineer"
    }
]

# Convert JSON data to DataFrame
df = pd.DataFrame(data)
df
```

```python
df.to_json() #to convert data to json but the problem here is it is taking index as a record too 
```

```python
df.to_json(orient='records') #json creates a complete record instead of taking the index number as a record
```

#  Reading HTML content for webscraping for tables

```python
url = 'https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/'
df = pd.read_html(url) #reading the failed bank data from the internet
df
```

```python
type(df)
```

```python
df[0] #Since the pd.read_html() function returns a list of DataFrames, df[0] selects the first DataFrame from the list.
```

```python
mob_url = 'https://en.wikipedia.org/wiki/Mobile_country_code'
df = pd.read_html(mob_url, match='Country') #parameter is used to filter the tables based on whether they contain the specified string ('Country' in this case). If no match is found, an empty list will be returned.
df[0]
```
