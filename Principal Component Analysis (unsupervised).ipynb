```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

The dataset is present in the form of dictionaries total rows are 569 and columns are 30

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])  #to check the decription of dataset
```

```python
#converting the data into the dataframe

df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

# PCA Visualizations

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #we are converting the 30 features in dataset into just 2 features
```

```python
pca.fit(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape #successfully converted to 2 features only
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap = 'plasma')
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
```
